# Importing Librarires

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

# Loading the Hotel Booking Data

```python
hotel_df = pd.read_csv('hotel_booking.csv')
```

# Exploratory Data Analysis and Data Cleaning

```python
hotel_df.shape
```

```python
hotel_df.head()
```

```python
hotel_df.columns
```

The dataset contains detailed information about hotel bookings, with the following columns:

1. hotel: Type of hotel (e.g., Resort Hotel).
2. is_canceled: Whether the booking was canceled or not.
3. lead_time: Number of days that elapsed between the booking and the arrival date.
4. arrival_date_year, arrival_date_month, arrival_date_week_number, arrival_date_day_of_month: Details about the arrival date.
5. stays_in_weekend_nights, stays_in_week_nights: Duration of the stay in terms of weekend and weekday nights.
6. adults, children, babies: Information about the guests.
7. meal: Type of meal booked.
8. country: Country of origin of the guest.
9. market_segment, distribution_channel: Information about the booking's market segment and distribution channel.
10. is_repeated_guest: Whether the guest has previously booked.
11. previous_cancellations, previous_bookings_not_canceled: Past booking behavior of the guest.
12. reserved_room_type, assigned_room_type: Details about the room type.
13. booking_changes: Number of changes/amendments made to the booking.
14. deposit_type: Type of deposit made.
15. agent, company: ID of the travel agency or company responsible for the booking.
16. days_in_waiting_list: Number of days the booking was in the waiting list before confirmation.
17. customer_type: Type of booking.
18. adr: Average daily rate.
19. required_car_parking_spaces: Number of car parking spaces required by the guest.
20. total_of_special_requests: Number of special requests made by the guest.
21. reservation_status, reservation_status_date: Status of the reservation and the date when the status was set.
22. Personal details like name, email, phone-number, and credit_card.

```python
# Removing the personal data of the coustomers

columns_to_drop = ['name', 'email', 'phone-number','credit_card']
hotel_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
hotel_df.info()
```

```python
# changing the datatype of reservation_status_date
hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'])
```

```python
hotel_df.describe(include='object')
```

```python
for col in hotel_df.describe(include = 'object').columns:
    print(col)
    print(hotel_df[col].unique())
    print('_'*50)
```

```python
# Checking for missing values in the dataset
hotel_df.isnull().sum()
```

```python
# Removing the agent and company column as it has largest missing data

columns_to_drop = ['company', 'agent']
hotel_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# Droping the missing value rows
hotel_df.dropna(inplce = True)
```

```python
# Stats of numerical columns
hotel_df.describe()
```

```python
hotel_df['adr'].plot(kind = 'box')
```

```python
# Removing the outlier from adr
hotel_df = hotel_df[hotel_df['adr']<5000]
```

# Data Analysis and Visualizations

## Booking Cancellation Analysis

```python
cancelled_perc = hotel_df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
plt.figure(figsize = (8,6))
plt.title('Reservation status count',weight='bold')
plt.bar(['Not cancelled','Cancelled'],hotel_df['is_canceled'].value_counts(),edgecolor = 'k' , width =0.7)
# Making x-axis tick labels bold
plt.xticks(weight='bold')
plt.show()
```

The accompanying bar graph shows the percentage of reservations that are cancelled and those that are not. It is obvious that there are still a significant number of reservations that have not been canceled.There are still 37% of clients who cancelled their reservation, which has a significant impact on the hotels' earnings.

```python
plt.figure(figsize=(8,6))
ax1 = sns.countplot(x='hotel',hue='is_canceled',data=hotel_df, palette='Blues')
lengend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(handles=lengend_labels,bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotels",size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

In comparison to resort hotels, city hotels have more bookings. It's possible that restort hotels are more expensive than those in cities.

```python
resort_hotel = hotel_df[hotel_df['hotel'] == 'Resort Hotel' ]
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = hotel_df[hotel_df['hotel'] == 'City Hotel' ]
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel' , fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

The line graph above shows that, on certain days, the average daily rate for a city hotel is less than that of a resort hotel, and on other days, it is even lesss. It goes without saying that weekends and holidays may see a rise in resort hotels rates.

```python
hotel_df['month'] = hotel_df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x='month' , hue ='is_canceled',data=hotel_df)
lengend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(handles=lengend_labels,bbox_to_anchor=(1,1))
plt.title("Reservation status per month",size=20)
plt.xlabel('number of reservations')
plt.ylabel('month')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize=30)
sns.barplot('month', 'adr', data = hotel_df[hotel_df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index())
plt.show()
```

```python
cancelled_data = hotel_df[hotel_df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (10,10))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country , autopct = '%2f',labels = top_10_country.index)
plt.show()
```

```python
hotel_df['market_segment'].value_counts()
```

```python
hotel_df['market_segment'].value_counts(normalize = True)
```

```python
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date' , inplace=True)

not_cancelled_data = hotel_df[hotel_df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date' , inplace=True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```

##  Guest Nationality Analysis

```python
# Grouping by country and counting the number of bookings
country_counts = hotel_df['country'].value_counts()

# Identifying the top 10 countries with the highest number of bookings
top_countries = country_counts.head(10)

# Visualizing the top countries
import matplotlib.pyplot as plt
top_countries.plot(kind='bar', figsize=(12,7))
plt.title('Top 10 Countries by Number of Bookings')
plt.ylabel('Number of Bookings')
plt.xlabel('Country')
plt.show()
```

- By analyzing the top countries from which guests originate, hotels can tailor their marketing strategies and offers.
- If a significant proportion of guests come from specific countries, the hotel could consider offering cultural events, food, or amenities specific to those nationalities to enhance their stay.
- This data can also be used for targeted advertising in specific regions or countries

## Duration of Stay

```python
# Creating a new column for total stay duration
hotel_df['total_stay'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']

# Visualizing the distribution of total stay durations
hotel_df['total_stay'].hist(bins=30, figsize=(12,7))
plt.title('Distribution of Duration of Stay')
plt.ylabel('Number of Bookings')
plt.xlabel('Duration (Nights)')
plt.show()
```

- Understanding the typical duration guests stay at the hotel helps in managing inventory, staffing, and other logistical aspects.
- For instance, if most guests typically stay for the weekend, the hotel can offer special weekend packages or events.
- Conversely, if the majority of the stays are during weekdays, it might suggest a business clientele, and the hotel can tailor its amenities and services accordingly.

## Special Requests Analysis

```python
# Grouping by number of special requests and calculating the cancellation rate
special_requests_group = hotel_df.groupby('total_of_special_requests')['is_canceled'].mean()

# Visualizing the impact of special requests on cancellation
special_requests_group.plot(kind='bar', figsize=(12,7))
plt.title('Cancellation Rate by Number of Special Requests')
plt.ylabel('Cancellation Rate')
plt.xlabel('Number of Special Requests')
plt.show()
```

- A correlation between the number of special requests and cancellations can provide insights into guest behavior. If cancellations increase with more special requests, it might suggest that guests aren't confident that the hotel can accommodate their needs.
- Proactively addressing common special requests in the booking process or ensuring timely communication with guests can potentially reduce cancellations.

## Repeated Guests

```python
# Calculating the proportion of repeated guests
repeated_guest_ratio = hotel_df['is_repeated_guest'].mean()
print(f"Proportion of Repeated Guests: {repeated_guest_ratio:.2%}")
```

- A high proportion of repeat guests indicates strong guest loyalty and satisfaction with the hotel's services. Such guests can be targeted with loyalty programs or exclusive offers.
- Conversely, a low proportion of repeated guests might suggest the need to improve guest experiences or conduct surveys to understand areas of improvement.

## Booking Preferences Analysis

```python
# Analyzing most preferred room types
room_type_counts = hotel_df['reserved_room_type'].value_counts()

# Visualizing the preferred room types
room_type_counts.plot(kind='bar', figsize=(12,7))
plt.title('Most Preferred Room Types')
plt.ylabel('Number of Bookings')
plt.xlabel('Room Type')
plt.show()
```

```python
# Analyzing meal preferences
meal_counts = hotel_df['meal'].value_counts()

# Visualizing the meal preferences
meal_counts.plot(kind='bar', figsize=(12,7))
plt.title('Meal Preferences')
plt.ylabel('Number of Bookings')
plt.xlabel('Meal Type')
plt.show()
```

- The most popular room types indicate where the hotel's primary revenue comes from and can guide room renovation or upgrade decisions.
- If certain room types have low bookings but occupy significant space, the hotel might consider repurposing or redesigning those rooms.
- Understanding the most popular meal plans can help in inventory management for the kitchen and optimize staffing needs. For instance, if the majority prefer bed and breakfast, it indicates the importance of providing a high-quality breakfast service.

## Monthly Analysis

```python
# Grouping by month to analyze monthly trends
monthly_counts = hotel_df['arrival_date_month'].value_counts().reindex(
    ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
)

# Visualizing monthly bookings
monthly_counts.plot(kind='line', marker='o', figsize=(12,7))
plt.title('Monthly Bookings Trend')
plt.ylabel('Number of Bookings')
plt.xlabel('Month')
plt.grid(True)
plt.show()
```

- The monthly trends show which months have the highest bookings, indicating peak seasons. This can help in price optimization, staffing, and inventory management.
- Identifying off-peak months can lead to targeted promotions or discounts to boost bookings during these periods. It's also an ideal time for renovations or maintenance activities.

## Market Segment Analysis

```python
# Analyzing market segments
market_segment_counts = hotel_df['market_segment'].value_counts()

# Visualizing the market segments
market_segment_counts.plot(kind='bar', figsize=(12,7))
plt.title('Bookings by Market Segment')
plt.ylabel('Number of Bookings')
plt.xlabel('Market Segment')
plt.show()
```

- Knowing which market segments are most prevalent helps in targeted marketing. For instance, if a significant portion comes from travel agents, strengthening partnerships or offering commissions might boost bookings.
- Conversely, if direct bookings are low, the hotel might look into improving its website or online booking system for a better user experience.

## Waiting List Analysis

```python
# Filtering bookings that were put in waiting list
waiting_list = hotel_df[hotel_df['days_in_waiting_list'] > 0]

# Status of bookings that were in waiting list
waiting_list_status = waiting_list['reservation_status'].value_counts()

# Visualizing the status of bookings that were in waiting list
waiting_list_status.plot(kind='bar', figsize=(12,7))
plt.title('Reservation Status of Bookings in Waiting List')
plt.ylabel('Number of Bookings')
plt.xlabel('Reservation Status')
plt.show()
```

- A high number of bookings on the waiting list that eventually get canceled might indicate overbooking practices or inefficiencies in managing room availability.
- If a significant portion of the waiting list bookings get confirmed, it shows that the waiting list system is effective. However, the hotel might still want to explore why so many bookings end up on the waiting list in the first place

# General Insights
